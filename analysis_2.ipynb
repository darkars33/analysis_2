```python
ratecard_df['Weight Slab'] = ratecard_df['Weight Slab'].astype(float)
ratecard_df
```

```python
total_A,Count_A = 0,0
total_B,Count_B = 0,0
total_C,Count_C = 0,0
total_D,Count_D = 0,0
total_E,Count_E = 0,0
shipsy_count = 0
```

### Calculating base cost

```python
for index, data in Data_Dump.iterrows():
    if data['Courier']=='Shipsy':
        shipsy_count += 1
        if data['Zone']=='A':
            Count_A += 1
            total_A += 150
        elif data['Zone']=='B':
            Count_B += 1
            total_B += 150
        elif data['Zone']=='C':
            Count_C += 1
            total_C += 150
        elif data['Zone']=='D':
            Count_D += 1
            total_D += 150
        elif data['Zone']=='E':
            Count_E += 1
            total_E += 150

    
    else:
        for index_2, rate in ratecard_df.iterrows():
            if data['Direction']==rate['Direction'] and data['Courier Slab'] == rate['Weight Slab']  and rate['Freight Type']=='Base':
                if data['Zone']=='A':
                    Count_A += 1
                    total_A += rate['A']
                elif data['Zone']=='B':
                    Count_B += 1
                    total_B += rate['B']
                elif data['Zone']=='C':
                    Count_C += 1
                    total_C += rate['C']
                elif data['Zone']=='D':
                    Count_D += 1
                    total_D += rate['D']
                elif data['Zone']=='E':
                    Count_E += 1
                    total_E += rate['E']
```

```python
Count_A+Count_B+Count_C+Count_D+Count_E
```

### Including cost for Additional weight

```python
for index, data in Data_Dump.iterrows():
    if data['Courier']=='Shipsy':
        continue
    
    else:
        for index_2, rate in ratecard_df.iterrows():
            if data['Direction']==rate['Direction'] and data['Courier Slab'] == rate['Weight Slab']  and rate['Freight Type']!='Base':
                if data['Zone']=='A':
                    total_A += rate['A']*data['Multiplier']
                    
                elif data['Zone']=='B':
                    total_B += rate['B']*data['Multiplier']
                    
                elif data['Zone']=='C':
                    total_C += rate['C']*data['Multiplier']
                    
                elif data['Zone']=='D':
                    total_D += rate['D']*data['Multiplier']
                    
                elif data['Zone']=='E':
                    total_E += rate['E']*data['Multiplier']
```

## Average shipping cost

```python
print("Average cost for Zone A is: ",total_A/Count_A)
print("Average cost for Zone B is: ",total_B/Count_B)
print("Average cost for Zone C is: ",total_C/Count_C)
print("Average cost for Zone D is: ",total_D/Count_D)
print("Average cost for Zone E is: ",total_E/Count_E)
```
